```python
import folium
import pandas as pd

taxi = pd.read_csv("./택시승차대 현황.csv", encoding = 'cp949')
```

```python
loc = taxi.loc[:,['위도','경도']]
```

```python
m = folium.Map(location=[37.547647896577175, 126.94245451708481], zoom_start=16)
```

```python
for i in range(0,len(loc)):
    y = loc.위도[i]
    x = loc.경도[i]
    folium.Marker([y,x], icon = folium.Icon(color = 'purple', icon = 'fa-solid fa-train-subway', prefix='fa-solid')).add_to(m)
    
m
```
