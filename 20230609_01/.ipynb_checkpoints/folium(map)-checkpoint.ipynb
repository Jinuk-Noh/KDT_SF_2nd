```python
import folium

m = folium.Map(location=[37.547647896577175, 126.94245451708481], zoom_start=16)

#layer 추가
folium.TileLayer('Stamen Terrain').add_to(m)
folium.LayerControl().add_to(m)

#마커 추가

#folium.Marker([위도, 경로], popup='', tooltip = '').add_to(m)
folium.Marker([37.547647896577175, 126.94245451708481], popup='subway', tooltip= '대흥역', icon = folium.Icon(color = 'purple', icon = 'fa-solid fa-train-subway', prefix='fa-solid')).add_to(m)
folium.Marker([37.544214528707485, 126.95121262987305], popup='subway', tooltip= '공덕역', icon = folium.Icon(icon = 'fa-solid fa-star', prefix='fa-solid')).add_to(m)

m.add_child(folium.ClickForMarker(popup= 'Test'))

m
```
